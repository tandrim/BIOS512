```r
library('tidyverse')
wind_data = read_csv('https://raw.githubusercontent.com/rfordatascience/tidytuesday/refs/heads/master/data/2020/2020-10-27/wind-turbine.csv')
head(wind_data)
```

Each row in the table represents a single turbine.

Make a bar chart ðŸ“Š **sorted by frequency** of turbine count per province territory. (Hint: `fct_infreq` will sort your factor levels by frequency for subsequent plotting)

```r
df = wind_data %>% 
    mutate(province_territory = fct_infreq(province_territory)) %>% 
    group_by(province_territory) %>% 
    summarize(count = n())

ggplot(df, aes(y= province_territory, x= count)) + 
    geom_col() + labs(x = 'Frequency', y= 'Province/Territory')
```

Make a bar chart **sorted by frequency** of turbine count per manufacturer. **Group the low frequency manufacturers into an `Other` category**.

**Bonus:** Can you force the `Other` category to be the last bar? 

```r
df_lumped = wind_data %>%
    mutate(manufacturer = fct_lump_min(manufacturer, min = 900)) %>%
    mutate(manufacturer = fct_relevel(manufacturer, "Other", after = Inf)) %>%
    mutate(manufacturer = manufacturer %>% fct_infreq() %>% fct_rev()) %>%
    group_by(manufacturer) %>%
    summarize(count = n())

head(df_lumped)

ggplot(df_lumped, aes(x= count, y= manufacturer)) + geom_col()
```
